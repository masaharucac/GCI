```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
dummied = pd.get_dummies(df[['Player_Type','Position_Type','Position']], drop_first=True)
df = pd.concat([df.drop(columns=['Player_Type','Position_Type','Position']), dummied], axis=1)
df = df.drop(columns=['Id','School','Position_Type_other_special', 'Position_LS', 'Position_DB'])
df
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(df.drop(columns='Drafted'), df['Drafted'], random_state=0, test_size=0.2)
```

```python
X_train_val.isna().sum()
```

```python
X_train_val['Age'] = X_train_val['Age'].fillna(X_train_val['Age'].mean())
X_train_val['Sprint_40yd'] = X_train_val['Sprint_40yd'].fillna(X_train_val['Sprint_40yd'].mean())
X_train_val['Vertical_Jump'] = X_train_val['Vertical_Jump'].fillna(X_train_val['Vertical_Jump'].mean())
X_train_val['Bench_Press_Reps'] = X_train_val['Bench_Press_Reps'].fillna(X_train_val['Bench_Press_Reps'].mean())
X_train_val['Broad_Jump'] = X_train_val['Broad_Jump'].fillna(X_train_val['Broad_Jump'].mean())
X_train_val['Agility_3cone'] = X_train_val['Agility_3cone'].fillna(X_train_val['Agility_3cone'].mean())
X_train_val['Shuttle'] = X_train_val['Shuttle'].fillna(X_train_val['Shuttle'].mean())
X_train_val.isna().sum()
```

```python
X_train_val
```

```python
pf = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)

X_train_val_poly = pf.fit_transform(X_train_val[['Age', 'Height', 'Weight', 'Sprint_40yd', 'Vertical_Jump', 'Bench_Press_Reps', 'Broad_Jump', 'Agility_3cone', 'Shuttle']])
X_train_val_poly = pd.DataFrame(X_train_val_poly, columns=pf.get_feature_names_out(X_train_val[['Age', 'Height', 'Weight', 'Sprint_40yd', 'Vertical_Jump', 'Bench_Press_Reps', 'Broad_Jump', 'Agility_3cone', 'Shuttle']].columns))
X_train_val_poly
```

```python

x_train_drafted_columns = X_train_val.drop(columns=['Age', 'Height', 'Weight', 'Sprint_40yd', 'Vertical_Jump', 'Bench_Press_Reps', 'Broad_Jump', 'Agility_3cone', 'Shuttle'])
# インデックスをリセット
X_train_val_poly = X_train_val_poly.reset_index(drop=True)
x_train_drafted_columns = x_train_drafted_columns.reset_index(drop=True)

# データフレームを結合
X_train_val1 = pd.concat([X_train_val_poly, x_train_drafted_columns], axis=1)

# 結果を確認
print(f"X_train_val1 shape: {X_train_val1.shape}")
```

```python
train_X, val_X, train_y, val_y =train_test_split(X_train_val1, y_train_val, random_state=0, test_size=0.2)
forest = RandomForestClassifier(n_estimators=100, random_state=2, max_depth=8)
forest.fit(train_X, train_y)
print(f"Validation Score: {forest.score(val_X, val_y)}")
print(f"Training Score: {forest.score(train_X, train_y)}\n")

from sklearn.metrics import roc_auc_score
print(f'ROC: {roc_auc_score(forest.predict(val_X), val_y)}')
```
